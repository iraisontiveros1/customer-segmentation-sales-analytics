```python

# 1. Import necessary libraries

import pandas as pd


# 2. Load the raw dataset from /data/raw/

df = pd.read_csv("../data/raw/customer_segmentation_dataset.csv")


# 3. Show the first rows of the dataset
#    This confirms the file loaded correctly

df.head()
```

```python
# CLEANING
# Remove duplicate rows
df = df.drop_duplicates()

# Remove rows where every column is empty
df = df.dropna(how="all")

# Show the first rows again
df.head()
```

```python
# Create a new column representing total sales per order
df["sales_amount"] = df["quantity"] * df["unit_price"]

# Show the first rows to confirm the new column was added
df.head()
```

```python
# Convert the order_date column to a proper datetime format
df["order_date"] = pd.to_datetime(df["order_date"])

# Check the data types to confirm the conversion
df.dtypes
```

```python
# Save the cleaned dataset into the processed folder
df.to_csv("../data/processed/customer_segmentation_clean.csv", index=False)
# Confirm save
"data_clean table saved successfully"
```
